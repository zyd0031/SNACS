```python
medium = nx.read_edgelist('medium.tsv', delimiter='\t', create_using=nx.DiGraph())
large = nx.read_edgelist('large.tsv', delimiter='\t', create_using=nx.DiGraph())
```

```python
medium_in_degree = list(dict(medium.in_degree()).values())
medium_out_degree = list(dict(medium.out_degree()).values())

large_in_degree = list(dict(large.in_degree()).values())
large_out_degree = list(dict(large.out_degree()).values())
```

```python
medium_in_degree_before_16 = [x for x in medium_in_degree if x <= 16]
medium_in_degree_after_16 = [x for x in medium_in_degree if x > 16]
```

```python
fig = plt.figure(layout = 'constrained')
axs = fig.subplot_mosaic([
    ["Medium's indegree distribution", "Medium's indegree distribution"],
    ["Medium's indegree distribution(before 16)", "Medium's indegree distribution(after 16)"]
])

axs["Medium's indegree distribution"].set_title("Medium's Indegree Distribution")
sns.kdeplot(medium_in_degree, ax = axs["Medium's indegree distribution"])

axs["Medium's indegree distribution(before 16)"].set_title(r"$Indegree \leq 16$")
axs["Medium's indegree distribution(before 16)"].hist(medium_in_degree_before_16, bins = 17)
axs["Medium's indegree distribution(before 16)"].set_ylabel("Frequency")

axs["Medium's indegree distribution(after 16)"].set_title(r"$Indegree > 16$")
axs["Medium's indegree distribution(after 16)"].hist(medium_in_degree_after_16, bins = 20)

fig.text(0.55, -0.05, "Indegree", ha = "center", fontsize = 12)
```

```python
medium_out_degree_before_16 = [x for x in medium_out_degree if x <= 16]
medium_out_degree_after_16 = [x for x in medium_out_degree if x > 16]

fig = plt.figure(layout = 'constrained')
axs = fig.subplot_mosaic([
    ["Medium's outdegree distribution", "Medium's outdegree distribution"],
    ["Medium's outdegree distribution(before 16)", "Medium's outdegree distribution(after 16)"]
])

axs["Medium's outdegree distribution"].set_title("Medium's Outdegree Distribution")
sns.kdeplot(medium_out_degree, ax = axs["Medium's outdegree distribution"])

axs["Medium's outdegree distribution(before 16)"].set_title(r"$Outdegree \leq 16$")
axs["Medium's outdegree distribution(before 16)"].hist(medium_out_degree_before_16, bins = 17)
axs["Medium's outdegree distribution(before 16)"].set_ylabel("Frequency")

axs["Medium's outdegree distribution(after 16)"].set_title(r"$Outdegree > 16$")
axs["Medium's outdegree distribution(after 16)"].hist(medium_out_degree_after_16, bins = 20)

fig.text(0.55, -0.05, "Outdegree", ha = "center", fontsize = 12)
```

```python
large_in_degree_before_17 = [x for x in large_in_degree if x <= 17]
large_in_degree_between_17_and_63 = [x for x in large_in_degree if  17 < x <= 63]
large_in_degree_after_63 = [x for x in large_in_degree if x > 63]

fig = plt.figure(layout = 'constrained')
axs = fig.subplot_mosaic([
    ["Large's indegree distribution", "Large's indegree distribution", "Large's indegree distribution"],
    ["Large's indegree distribution(before 17)", "Large's indegree distribution(between 17 and 63)", "Large's indegree distribution(after 63)"]
])

axs["Large's indegree distribution"].set_title("Large's Indegree Distribution")
sns.kdeplot(large_in_degree, ax = axs["Large's indegree distribution"])

axs["Large's indegree distribution(before 17)"].set_title(r"$Indegree \leq 17$")
axs["Large's indegree distribution(before 17)"].hist(large_in_degree_before_17, bins = 18)
axs["Large's indegree distribution(before 17)"].set_ylabel("Frequency")


axs["Large's indegree distribution(between 17 and 63)"].set_title(r"$17 < Indegree \leq 63$")
axs["Large's indegree distribution(between 17 and 63)"].hist(large_in_degree_between_17_and_63, bins = 20)

axs["Large's indegree distribution(after 63)"].set_title(r"$Indegree > 63$")
axs["Large's indegree distribution(after 63)"].hist(large_in_degree_after_63, bins = 30)

fig.text(0.55, -0.05, "Indegree", ha = "center", fontsize = 12)
```

```python
large_out_degree_before_17 = [x for x in large_out_degree if x <= 17]
large_out_degree_between_17_and_47 = [x for x in large_out_degree if  17 < x <= 47]
large_out_degree_after_47 = [x for x in large_out_degree if x > 47]

fig = plt.figure(layout = 'constrained')
axs = fig.subplot_mosaic([
    ["Large's outdegree distribution", "Large's outdegree distribution", "Large's outdegree distribution"],
    ["Large's outdegree distribution(before 17)", "Large's outdegree distribution(between 17 and 47)", "Large's outdegree distribution(after 47)"]
])

axs["Large's outdegree distribution"].set_title("Large's Outdegree Distribution")
sns.kdeplot(large_out_degree, ax = axs["Large's outdegree distribution"])

axs["Large's outdegree distribution(before 17)"].set_title(r"$Indegree \leq 17$")
axs["Large's outdegree distribution(before 17)"].hist(large_in_degree_before_17, bins = 18)
axs["Large's outdegree distribution(before 17)"].set_ylabel("Frequency")


axs["Large's outdegree distribution(between 17 and 47)"].set_title(r"$17 < Indegree \leq 47$")
axs["Large's outdegree distribution(between 17 and 47)"].hist(large_in_degree_between_17_and_63, bins = 20)

axs["Large's outdegree distribution(after 47)"].set_title(r"$Indegree > 47$")
axs["Large's outdegree distribution(after 47)"].hist(large_in_degree_after_63, bins = 30)

fig.text(0.55, -0.05, "Outdegree", ha = "center", fontsize = 12)
```

```python
medium_distance = []

for source, dictionary in nx.all_pairs_shortest_path_length(medium_largest_weakly_graph):
    distances = list(dictionary.values())
    distances.remove(0)
    medium_distance.extend(distances)
```

```python
import numpy as np
np.mean(medium_distance)
```

```python
from collections import Counter
import numpy as np

medium_distance = Counter(medium_distance)
distances = list(medium_distance.keys())
frequencies = list(medium_distance.values())

fig, ax = plt.subplots()

ax.bar(distances, np.log(frequencies), width = 1, edgecolor = 'grey', linewidth = 0.2)
plt.xlabel("Distance")
plt.ylabel("Log(Frequency)")
plt.title("Distribution Histogram (Log Scale)")
plt.suptitle("The Largest Weakly Connected Component of the Medium Graph")
plt.yscale("log")
```

```python
large_distance = []

for source, dictionary in nx.all_pairs_shortest_path_length(large_largest_weakly_graph):
    distances = list(dictionary.values())
    distances.remove(0)
    large_distance.extend(distances)
```

```python
from collections import Counter
import numpy as np

large_distance = Counter(large_distance)
distances = list(large_distance.keys())
frequencies = list(large_distance.values())
```

```python
fig, ax = plt.subplots()

ax.bar(distances, np.log(frequencies), width = 1, edgecolor = 'grey', linewidth = 0.2)
plt.xlabel("Distance")
plt.ylabel("Log(Frequency)")
plt.title("Distribution Histogram (Log Scale)")
plt.suptitle("The Largest Weakly Connected Component of the Large Graph")
plt.yscale("log")
```
